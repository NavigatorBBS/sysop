# MaxBot Basic Usage Example

This notebook demonstrates the basic usage of MaxBot with GitHub Copilot SDK.

## Setup

First, load environment variables and the MaxBot extension:

```python
from dotenv import load_dotenv
load_dotenv()

%load_ext maxbot
```

## Basic Chat

## Code Analysis

Ask MaxBot to analyze Python code:

```python
code = '''
import pandas as pd

df = pd.read_csv('data.csv')
df = df[df['value'] > 0]
result = df.groupby('category').sum()
'''

response = await agent.analyze_code(code, context="data aggregation")
response
```

## Multi-turn Conversation

MaxBot maintains conversation history:

```python
# First question
response = await agent.chat("What's the difference between .loc and .iloc in pandas?")
response
```

```python
# Follow-up question (uses context from previous question)
response = await agent.chat("Can you show me an example of each?")
response
```

## View Conversation History

```python
messages = await agent.get_messages()
print(f"Conversation has {len(messages)} messages")
```

## Clear History

Start a fresh conversation:

```python
await agent.clear_history()
print("History cleared!")
```

## Plain Text Response

Get responses without markdown formatting:

```python
plain_text = await agent.chat("Explain list comprehensions", as_markdown=False)
print(plain_text)
```

## Cleanup

When you're done, clean up resources:

```python
await agent.cleanup()
print("✅ Resources cleaned up")
```
